```python
data6['N_Gender'] = [0 if data6.loc[i,'Gender'] == 'Female' 
                    else 1 for i in range(len(data6))]

data6['N_Department'] = [0 if data6.loc[i,'Department'] == 'CSE' 
                        else 1 if data6.loc[i,'Department'] == 'Chemical'
                        else 2 if data6.loc[i,'Department'] == 'Civil'
                        else 3 if data6.loc[i,'Department'] == 'EE'
                        else 4 if data6.loc[i,'Department'] == 'GE'
                        else 5 if data6.loc[i,'Department'] == 'ME1'
                        else 6 for i in range(len(data6))]

data6['N_AbsentDays'] = [0 if data6.loc[i,'AbsentDays'] == '0-5'
                        else 1 if data6.loc[i,'AbsentDays'] == '6-10'
                        else 2 if data6.loc[i,'AbsentDays'] == '11-15'
                        else 3 for i in range(len(data6))]

data6['N_GuardianResponsible'] = [0 if data6.loc[i,'GuardianResponsible'] == 'Mother'
                        else 1 if data6.loc[i,'GuardianResponsible'] == 'Father'
                        else 2 if data6.loc[i,'GuardianResponsible'] == 'Both'
                        else 3 for i in range(len(data6))]

data6['N_EducationOfFather'] = [0 if data6.loc[i,'EducationOfFather'] == 'Below Bachelors'
                        else 1 if data6.loc[i,'EducationOfFather'] == 'Bachelors'
                        else 2 for i in range(len(data6))]

data6['N_EducationOfMother'] = [0 if data6.loc[i,'EducationOfMother'] == 'Below Bachelors'
                        else 1 if data6.loc[i,'EducationOfMother'] == 'Bachelors'
                        else 2 for i in range(len(data6))]

data6['N_TotalAnnualIncome'] = [0 if data6.loc[i,'TotalAnnualIncome'] == '0-100000'
                        else 1 if data6.loc[i,'TotalAnnualIncome'] == '100000-500000'
                        else 2 if data6.loc[i,'TotalAnnualIncome'] == '500000-1000000'
                        else 3 for i in range(len(data6))]

data6['N_ViewResources'] = [0 if data6.loc[i,'ViewResources'] == 'Never'
                        else 1 if data6.loc[i,'ViewResources'] == 'Very Rarely'
                        else 2 if data6.loc[i,'ViewResources'] == 'Sometimes'
                        else 3 for i in range(len(data6))]

data6['N_GroupDiscussion'] = [0 if data6.loc[i,'GroupDiscussion'] == 'Never'
                        else 1 if data6.loc[i,'GroupDiscussion'] == 'Very Rarely'
                        else 2 if data6.loc[i,'GroupDiscussion'] == 'Sometimes'
                        else 3 for i in range(len(data6))]

data6['N_GroupDiscussion'] = [0 if data6.loc[i,'GroupDiscussion'] == 'Never'
                        else 1 if data6.loc[i,'GroupDiscussion'] == 'Very Rarely'
                        else 2 if data6.loc[i,'GroupDiscussion'] == 'Sometimes'
                        else 3 for i in range(len(data6))]

data6['N_ExtraActivities'] = [0 if data6.loc[i,'ExtraActivities'] == 'Never'
                        else 1 if data6.loc[i,'ExtraActivities'] == 'Very Rarely'
                        else 2 if data6.loc[i,'ExtraActivities'] == 'Sometimes'
                        else 3 for i in range(len(data6))]
data6['Grade'] = [0 if data.loc[i,'GPA'] < 2 
                 else 1 for i in range(len(data))]
```

```python
a,x = plt.subplots(figsize=(16, 10))
sns.heatmap(data6.corr(), annot=True, linecolor='white', square = True, 
            linewidths= 0.1, vmax = 1, fmt= '.3f', cmap = 'RdBu',ax=x)
plt.title('Correlation between different features', y=1.03, size=18)
plt.show()
```

```python
f,ax=plt.subplots(2,2,figsize=(14, 12))
colors1 = ['lightblue', 'pink']
data6['Gender'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0,0],shadow=False, 
                                       explode=(0, 0.15), colors = colors1)
ax[0,0].set_title('Gender')
ax[0,0].set_ylabel('')

sex = data6.Gender.unique()
mean_sex=[sum(data6[data6.Gender == i].GPA)/float(len(data6[data6.Gender == i])) 
                   for i in sex]
sns.barplot(x=sex, y=mean_sex, ax=ax[0,1], facecolor=(0, 0, 0, 0), linewidth=8, 
              edgecolor=sns.color_palette("dark", 5))
ax[0,1].set_title('Gender vs GPA')


colors2 = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink']
data6['BirthDistrict'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[1,0],shadow=False, colors = colors2)

ax[1,0].set_ylabel('')
ax[1,0].set_title('BirthDistrict')

nationality = data6.BirthDistrict.unique()
mean_nationality=[sum(data6[data6.BirthDistrict == i].GPA)/float(len(data6[data6.BirthDistrict == i])) 
                   for i in nationality]
sns.barplot(x=nationality, y=mean_nationality, ax=ax[1,1], facecolor=(0, 0, 0, 0), linewidth=6, 
              edgecolor=sns.color_palette("dark", 20))

ax[1,1].set_title('BirthDistrict vs GPA')
plt.xticks(rotation=70)
plt.show()
```
